```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#I'm going to work with the tips data set
tips = sns.load_dataset("tips")
print(tips.head(n=4))
```

```python
#Choose one variable and plot that variable four different ways.
plt.plot(tips['tip'], color='green')
plt.ylabel('Tip Amount ($)')
plt.xlabel('Tip Index')
plt.title('Plot 1a: Tip Data')
plt.show()
```

```python
ax = sns.boxplot(data=tips['tip'], color='green')  
plt.title('Plot 1b: Tips Boxplot')
ax.set(xlabel='', ylabel='Tip Amount ($)')
plt.show()
```

```python

plt.hist(tips['tip'], color='green') 
plt.xlabel('Tip Amount ($)')
plt.ylabel('Count of Tips')
plt.title('Plot 1c: Tips Histogram')
plt.show()
```

```python
plt.plot(tips['tip'].sort_values())

#not working as intended, alternative below
```

```python
ls = list()
n = 0
while n < len(tips['tip']):
    ls.append(n)
    n += 1

plt.scatter(x=ls, y=tips['tip'].sort_values(), color='green')
plt.xlabel('Rank')
plt.ylabel('Tip Amount ($)')
plt.title('Plot 1d: Ranked Tips')
plt.show()
```

```python
#Choose two continuous variables, and plot them three different ways.

g = sns.lmplot(y='tip', x='total_bill', data=tips, 
               fit_reg=True, scatter_kws={'alpha':0.4})
g.set_ylabels("Tip Amount ($)")
g.set_xlabels("Total Bill ($)")
plt.title('Plot 2a: Tip Amount by Total Bill')
plt.show()
```

```python
plt.hist(tips['tip'], color='green',  alpha=.5,label='Tip Amount') 
plt.hist(tips['total_bill'], color='blue', alpha=.5,label='Total Bill')
plt.xlabel('Tip Amount ($)')
plt.legend(loc='upper right')
plt.title('Plot 2b: Tip Amount and Total Bill Histogram')
plt.show()
```

```python
ax = sns.boxplot(data=tips['tip']/tips['total_bill'], color='blue')  
plt.title('Plot 2: Meal Tips Boxplot')
ax.set(xlabel='', ylabel='Tip Amount / Total Bill')
plt.title('Plot 2c: Tip Percent of Bill')
plt.show()
```

```python
#Choose one continuous variable and one categorical variable, and plot them six different ways.

plt.plot(tips.loc[tips['sex']=='Male','tip'])
plt.plot(tips.loc[tips['sex']=='Female','tip'])
plt.title('Plot 3a: Male and Female Tips')
plt.ylabel('Tip Amount ($)')

plt.show()
```

```python
sortedmale = tips.loc[tips['sex']=='Male','tip'].sort_values()
lenmale = list()
n = 0
while n < len(sortedmale):
    lenmale.append(n)
    n += 1

sortedfemale = tips.loc[tips['sex']=='Female','tip'].sort_values()
lenfemale = list()
n = 0
    
while n < len(sortedfemale):
    lenfemale.append(n)
    n += 1
    
plt.scatter(x=lenmale, y=sortedmale)
plt.scatter(x=lenfemale, y=sortedfemale)
plt.title('Plot 3b: Ranked Male and Female Tips')
plt.ylabel('Tip Amount ($)')
plt.show()
```

```python
sns.set(style="whitegrid")

g = sns.factorplot(x="sex", y="tip", data=tips,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Tip Amount ($)")
g.set_xlabels("Gender")
plt.title('Plot 3c: Meal Tips by Gender Pointplot')
plt.show()
```

```python
menlunchtip = tips.loc[(tips['sex']=='Male')&(tips['time']=='Lunch'),'tip']


plt.hist(tips.loc[tips['sex']=='Male','tip'],  alpha=.5,label='Men')  # alpha just controls the opacity
plt.hist(tips.loc[tips['sex']=='Female','tip'],  alpha=.5,label='Women')  # alpha just controls the opacity
plt.xlabel('Tip Amount ($)')
plt.legend(loc='upper right')
plt.title('Plot 3d: Meal Tips by Gender Hist.')
plt.show()
```

```python
ax = sns.boxplot(x='sex',y='tip',data=tips)  
plt.title('Plot 3e: Meal Tips by Gender Boxplot')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Tips')

plt.show()
```

```python
sns.set(style="white")

g = sns.lmplot(y='tip', 
               x='tip',
               hue='sex',
               data=tips, # Data.
               fit_reg=False,
               ci=False,
               scatter_kws={'alpha':0.4})
g.set_ylabels("Tip Amount ($)")
g.set_xlabels("Tip Amount ($)")
plt.suptitle('Plot 3f: Male and Female Tip Scatterplot')
plt.subplots_adjust(top=0.9)
plt.show()
```
